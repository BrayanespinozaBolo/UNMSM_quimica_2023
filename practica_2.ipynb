Actividad: Elaborar dataframes y comparar sus medidas de tendencia central para explicar las gráficas (Ref. Teoría cinética molecular de los gases)

boltz_maxw.PNG

```
import pandas as pd
import numpy as np
from tabulate import tabulate
N = 6.022e19
m = 0.028
k = 1.38e-23
T = [100, 300, 700]
temperaturas = []
velocidades_rms = []
# Calcular
for temp in T:
    v_rms = np.sqrt((3 * k * temp) / (m / N))
    temperaturas.append(temp)
    velocidades_rms.append(v_rms)

df = pd.DataFrame({
    'Temperatura (K)': temperaturas,
    'Velocidad cuadrática media (m/s)': velocidades_rms
})
media = df['Velocidad cuadrática media (m/s)'].mean()
mediana = df['Velocidad cuadrática media (m/s)'].median()

print(f'\nMedia de la velocidad cuadrática media: {media} m/s')
print(f'Mediana de la velocidad cuadrática media: {mediana} m/s')
```

```
import pandas as pd
import numpy as np
from tabulate import tabulate
k = 1.38e-23  # Constante de Boltzmann (J/K)
m_cl2 = 0.071  # Masa molar del cloro (kg/mol)
m_n2 = 0.028  # Masa molar del nitrógeno (kg/mol)
m_he = 0.004   # Masa molar del helio (kg/mol)
v_rms_cl2 = np.sqrt(3 * k * 300 / (m_cl2 / 6.022e23))
v_rms_n2 = np.sqrt(3 * k * 300 / (m_n2 / 6.022e23))
v_rms_he = np.sqrt(3 * k * 300 / (m_he / 6.022e23))
df = pd.DataFrame({
    'Molécula': ['Cl2', 'N2', 'He'],
    'Velocidad cuadrática media (m/s)': [v_rms_cl2, v_rms_n2, v_rms_he]
})
media = df['Velocidad cuadrática media (m/s)'].mean()
mediana = df['Velocidad cuadrática media (m/s)'].median()
print(f'\nMedia: {media} m/s')
print(f'Mediana: {mediana} m/s')
```
